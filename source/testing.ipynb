```python
import torch
import pandas as pd
from IPython.display import display

results_file_path = "/home/rohan/code/duel-sam/dueling-sam/source/outputs/20250702_180106_samTrue/results.pt"

results = torch.load(results_file_path)

data_for_df = {}
num_epochs = 0

for key, value in results.items():
    if isinstance(value, (list, torch.Tensor)):
        # Convert tensors to Python numbers if necessary
        data_for_df[key] = [v.item() if isinstance(v, torch.Tensor) else v for v in value]
        if not num_epochs:
            num_epochs = len(value)
    else:
        # Repeat hyperparameters for each epoch row
        data_for_df[key] = [value] * num_epochs

data_for_df['Epoch'] = list(range(1, num_epochs + 1))

df = pd.DataFrame(data_for_df)


column_order_preference = ['Epoch', 'train_loss', 'val_loss', 'test_loss',
                           'train_accuracy', 'val_accuracy', 'test_accuracy',
                           'learning_rate', 'batch_size', 'epochs'] 

final_columns = [col for col in column_order_preference if col in df.columns]
remaining_columns = [col for col in df.columns if col not in final_columns]
df = df[final_columns + remaining_columns]

print(f"Loaded and displaying results from: {results_file_path}\n")
display(df)
```
